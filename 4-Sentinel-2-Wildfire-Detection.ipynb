# Satellite Image Processing - Wildfire Detection

## Wildlfires

**Wildfires** are becoming more and more constant around the globe in the past recent years. **Wildfires** is the occurrence of fire out of control in any type of vegetation, and can be **caused** either by **human action** or by **nature's actions**, such as lightning.

Although this environmental phenomena may be important for vegetation renovation and agricultural practices for instance, it poses a **high risk** to our planet in several aspects, causing the **destruction of animal and plant life**, making the **soil poor**, being impossible for new planting. It can cause injuries to **animals** in the area and nearby regions, **destroying roads**, **bridges**, built-up areas in general, also causing **drastic changes** to the **ecosystem**, affecting reproduction and contributing to the spread of pests and diseases.

The **increased occurrence of  wildfires** and their **consequences** are also being related to **climate change**. In addition to the problems mentioned above, **wildfires** can **release** a huge amount of **greenhouse gases** to the **atmosphere**, which can facilitate the **rise of global temperatures**.

---

## Wildfires and Carbon emissions

**Wildfires** have contributed to an **increase in carbon dioxide (CO2) levels** recorded in the **Earth's atmosphere** since the measurement began more than 60 years ago, according to the UK Department of Meteorology (Met Office in January 2020).

**Predicted CO2 concentrations** over the next few years are **similar** to those in Earth's atmosphere between **3 and 5 million years ago**. At that time, **sea level** was **10 to 20 meters higher** than **today** and **temperatures** around **2°C and 3°C warmer**.

The increase in **greenhouse gas emissions** contributes to **global warming**, which can be seen as **one of the main factors** driving the increase in **wildfires** that we have seen recently.

Getting a concrete example, we can look into the numbers provided by the **Australian research institute Future Directions International - [FDI]( https://www.futuredirections.org.au/publication/wildfires-greenhouse-gas-emissions-and-climate-change/)**. This overview shows the significant wildfires contribution to carbon emissions:

* **Wildfires** can burn up to **120 million hectares**. The average is estimated at **30 million hectares per year**.<br>


* Usually, **wildfires** releases to the atmosphere a range of **5 to 30 tonnes of carbon per hectare**. In a **year** the total amount is estimated between **150 and 900 million tonnes of carbon**, and in particular situations **up to 600 and 3600 million tonnes of carbon a year**.<br>


* The numbers from the **industrial sector per year** in Australia can reach up to **155 million tonnes of carbon emission**.<br>

The point is that comparing **greenhouse gas emissions** from **industries** with **wildfires carbon emissions**, the **latter** can **surpass** the **former by up to 6 times**. ***This is huge***!

So, thinking in the **future** of planet **Earth**, there is a need to understand and study **wildfires** in general, and their related **carbon emissions**.

## Wildfires in the Artic

The **Arctic region** is being heavily affected by **wildfires** (vast areas of northern Siberia, Scandinavia, Alaska and Greenland), which may sound a little strange considering the characteristic cold climate of these regions.

**Fires** in the **Artic region** are usually caused by **lightning**, but lately the situation has been aggravated by **summer heat waves**. Temperatures are higher than average as a result of climate change. And sometimes the fire is so intense that the smoke can even be seen from **space!**

**Arctic fires** are common between **May** and **October**, and **wildfires** are a natural part of an ecosystem, offering some environmental benefits, according to the Alaska Centers website.

They **help** with the **growth of new grass and shrubs**, attracting rodents and partridges - and then foxes, minks and birds of prey. In addition, it allows herbs, tree sprouts and wild fruits to attract moose and bears, and allows for a **recycling of soil nutrients**.

But the **intensity of the flames**, as well as the **large area** they occupy, make the most recent episodes unusual.
**Dry land** and extremely **hotter** than average **temperatures**, combined with **lightning** and **strong winds**, caused the flames to **spread** aggressively. The **burning** was sustained by the **forest floor**, composed of **thawed**, **exposed** and **dry peat** - a soil type with **high carbon content**.

According to some scientists in the field, what we are seeing is **evidence** of the kind of **reaction** we should expect from a **warmer world**, where **increasing concentrations of greenhouse gases generate more warming**, leading to conditions that **release** even more carbon in the **atmosphere**.

**Siberian** recent cases can be seen in the following news: [CNN](https://edition.cnn.com/2020/09/03/world/arctic-wildfires-climate-intl/index.html).

## Spectral Indices - Normalized Burn Ratio (NBR)

The **Normalized Burn Ratio (NBR)** is a normalized difference index of the **reflectance** relative to the **NIR and SWIR** (Bands **8A** and **12** from **Sentinel-2**, respectively) bands of the satellite, in which it **highlights burnt areas** and **their severity**.  More details can be found in [UN-SPIDER KNOWLEDGE PORTAL](https://un-spider.org/advisory-support/recommended-practices/recommended-practice-burn-severity/in-detail/normalized-burn-ratio).

In this notebook, we will use **NBR** for **fire detection** study. We will first prepare a raster composite of a **fire event**, and then we will apply the well know formula of **NBR** to identify the **fire patches**. After, we will create two other **NBR** products, from **pre-** and **post- fire event**.

The equation for **NBR calculation** is presented below:

<img src='jpg/NBR_equation.jpg' />

**NBR** values range from **-1 to 1**, meaning values **closer to 1**, the **greater** the **presence of vegetation** or **unburned**, **slightly burned** areas. On the other hand, values **closer to -1**, the **greater** the **presence of bare soils and affected vegetation**, meaning **moderate** to **high burned** areas.

##  Spectral Indices - Difference Normalized Burn Ratio (dNBR) 

The **Difference Normalized Burn Ratio (dNBR)** is the calculation of the difference between **pre-** and **post-fire NBR** images. ### Burn Severity

According to **Key & Benson (2006)**, **burn severity** is the result or effects that **fire occurrences** produce on **ecosystems** in the **landscape**. The **dNBR** calculation quantifies and divide the **burn severity** into **different classes** (based on **USGS**). The higher the **dNBR** value, the greater the damage. [Humboldt State University](http://gsp.humboldt.edu/OLM/Courses/GSP_216_Online/lesson5-1/NBR.html):

|      ΔNBR        |       Burn Severity
|:----------------:|:----------------------------:|
|< -0.25	       |   High post-fire regrowth    |
|-0.25 to -0.1	   |   Low post-fire regrowth     |
|-0.1 to +0.1	   |          Unburned            |
|0.1 to 0.27	   |     Low-severity burn        |
|0.27 to 0.44	   |   Moderate-low severity burn |
|0.44 to 0.66	   |  Moderate-high severity burn |
|> 0.66	           |      High-severity burn      |



* Key, C.H. & Benson, N.C. (2006). FIREMON - Landscape assessment. Pages LA 1-55 *In*: Lutes, D.C., Keane, R.E., Caratti,J.F., Key, C.H., Benson, N.C., Sutherland, S., Gangi, L.J. (2006). FIREMON: Fire effects monitoring and inventory system. *General Technical Report RMRS-GTR-164-CD*. Fort Collins, CO: USDA Forest Service, Rocky Mountain Research Station. https://www.fs.fed.us/rm/pubs/rmrs_gtr164/rmrs_gtr164_13_land_assess.pdf.<br>

## Fire Detection using Sentinel-2 imagery

We can identify fire patches using **Sentinel-2** for example, getting the **active fire** as well as the **burnt scar**. For both examples we will need the **red**, **narrow nir** and **swir bands** from **Sentinel-2**. First, we will create a raster composite with a **total of four bands**, and then we will parse three bands out of the raster to visualize the active fire. After, we will parse them once again retrieving the necessary bands to calculate the **NBR spectral index**. 

In summary, the bands we need are:


                                                 -Raster composite with 4 bands-

|      Band number of our composite     |       Band number of Sentinel-2        |
|:------------------------------------: | :-------------------------------------:|
|               Band 1 - Red            |            Band 4 - Red                |
| Band 2 - Narror NIR (Near-infrared)   |     Band 8A - NIR (Near-infrared)      |
| Band 3 - SWIR (Short-wave infrared 1) | Band 11 - SWIR (Short-wave infrared 1) |
| Band 4 - SWIR (Short-wave infrared 2) | Band 12 - SWIR (Short-wave infrared 1) |

### ***Let's start coding and working on an interesting topic!***

```python
# importing necessary libraries

import rasterio
from rasterio import plot
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

### Accessing the images and creating a composite

Let's use the path directory where all bands we need are located:

```python
# Getting path directory
imagePath = "./data/"
```

We are going to load each band using the rasterio library. As explained above, in our fire detection example we will need **bands 4**, **8A**, **11** and **12** from **Sentinel-2**.

```python
# Using path directory to load each band
band4_wvp = rasterio.open(imagePath+'T51WVP_20200715T033539_B04_20m.jp2', driver='JP2OpenJPEG') #red
band8a_wvp = rasterio.open(imagePath+'T51WVP_20200715T033539_B8A_20m.jp2', driver='JP2OpenJPEG') #narrow nir
band11_wvp = rasterio.open(imagePath+'T51WVP_20200715T033539_B11_20m.jp2', driver='JP2OpenJPEG') #swir
band12_wvp = rasterio.open(imagePath+'T51WVP_20200715T033539_B12_20m.jp2', driver='JP2OpenJPEG') #swir
```

```python
# quick check on band metadata
band4_wvp.meta
```

### Reading Sentinel-2 image bands and writing the composite

Also using ***rasterio***, we will save to disk our four bands raster composite. We can choose one band to get the metadata we need, "attaching" this information to the new **tiff** file.

```python
#creating a new tiff composite
composite_wvp = rasterio.open(imagePath+'S2B_MSIL2A_20200715T033539_N0214_R061_T51WVP_4bands.tiff','w',driver='GTiff',
                         width=band4_wvp.width, height=band4_wvp.height,
                         count=4,
                         crs="+proj=utm +zone=49 +ellps=WGS84 +units=m +no_defs ",
                         transform=band4_wvp.transform,
                         dtype=band4_wvp.dtypes[0]
                         )
# parsing bands RGB
composite_wvp.write(band4_wvp.read(1),1) #red
composite_wvp.write(band8a_wvp.read(1),2) #narrow nir
composite_wvp.write(band11_wvp.read(1),3) #swir
composite_wvp.write(band12_wvp.read(1),4) #swir
composite_wvp.close()
```

There are different **band combinations** of **optical satellites** that we can use to display **active fires**. Another option would be **B12**, **B11** and **B8/B8A**.

In the current activity, we will use the first example **(B12, B11 and B4)**.  

```python
# opening the dataset with rasterio
fire_tiff = rasterio.open(imagePath+'S2B_MSIL2A_20200715T033539_N0214_R061_T51WVP_4bands.tiff',driver='Gtiff')
```

*Before using ***matplotlib*** here to display our **raster composite**, we are going to scale each band increasing the visual contrast of the image, and having a plot with higher visual quality. It is important to point out that this simple ***scaling math*** is only applied here for ***visual purposes*** and not to change the true value of each pixel and use these further in other processing activities.*

```python
# scaling raster values for display purposes
def scale(band):
    return band / 2000.0

# Loading each band into numpy arrays
band_red = scale(fire_tiff.read(1))
band_swir11 = scale(fire_tiff.read(3))
band_swir12 = scale(fire_tiff.read(4))
```

## Normalized Burn Ratio (NBR)

As we previously saw, the **NBR** spectral index is an important indicator to identify fire patches. It is optimal to detect **large fire zones**, but it can also work very well on **small wildfires**.

We will now calculate the NBR for the set of images we have shown above. We need only two bands from Sentinel-2, band 8A and band 12, nir and swir2 respectively.

---**Important reminder:**---<br>
**NBR values** range from **-1 to 1**. Negative values indicate stressed, dryed or burned vegetation, while positive values show healthier plants and/or unburned (or lightly burned) areas.  

In the last section we have already parsed **band 12** into a ***numpy** array, but we didn't do it for **band number 4**. So, let's load the latter into a ***numpy*** array and **scale** it like we did for the other bands.

```python
# Loading the nir band (band 4) into numpy array
band_nir = scale(fire_tiff.read(2))
```

Let's finally calculate the **NBR**:

```python
# allowing division by zero
np.seterr(divide='ignore', invalid='ignore')

nbr = (band_nir.astype(float) - band_swir12.astype(float)) / (band_nir + band_swir12)
```

Let's take a look at the range of values:

```python
# quick check on calculated NBR values
print(np.nanmin(nbr)) 
print(np.nanmax(nbr))
```

### Writing/Saving NBR image

As we are satisfied with our results, let's save this file as a **new tiff**. First, we will give the same **metadata** of our **input raster composite**.

```python
# getting the metadata of the input tiff
meta = fire_tiff.meta
print(meta)


# retrieving the dtype from our nbr
nbr_dtype = nbr.dtype
print(nbr_dtype)
```

```python
# updating dtype to float64
meta.update({"dtype": "float64"})
print(meta)
```

```python
# creating kwargs variable from meta
kwargs = meta


# updating the count value
kwargs.update(count=1)
```

```python
# saving new raster file
with rasterio.open(imagePath+'S2B_MSIL2A_20200715T033539_N0214_R061_T51WVP_nbr.tiff', 'w', **kwargs) as dst:
        dst.write(nbr, 1)
```

## Processing Difference Normalized Burn Ratio (dNBR)

Now is the time to make a **dNBR** product. The **dNBR** is the calculated **difference** between the **pre-fire image** and the **post-fire image**. 

As discussed earlier at the beggining of this notebook, the **dNBR results** indicate **burn severity** in the affected area. We can also use **dNBR** to contrast the most impacted regions against lightly burned or even unburned areas.

```python
# calculating dnbr
dnbr = nbr_pre - nbr_post
```

## Calculating burnt area of dNBR classes

```python
# importing necessary library
import math
```

We are defining a **function** to **reclassify** the **pixels** available in our **dnbr** product to integers switching our range of class values so that we can properly calculate the **area** of **each burnt severity class**. This **function** is **"normalizing"** the dataset values to a **higher level view**, otherwise the **float** values of the ***numpy array*** are constrained to a too **small range** and the results would not be representative.

```python
def reclassify(array):
    """
    This function reclassifies an array
    input:  array           array (n x m)    input array
    output: reclass         array (n x m)    reclassified array
    """
    reclass = np.zeros((array.shape[0],array.shape[1]))
    for i in range(0,array.shape[0]):
        for j in range(0,array.shape[1]):
            if math.isnan(array[i,j]):
                reclass[i,j] = np.nan
            elif array[i,j] < 0.1:
                reclass[i,j] = 1
            elif array[i,j] < 0.27:
                 reclass[i,j] = 2
            elif array[i,j] < 0.44:
                 reclass[i,j] = 3
            elif array[i,j] < 0.66:
                 reclass[i,j] = 4
            else:
                reclass[i,j] = 5
                
    return reclass
```

Calculating **burnt area size** of our **Siberian** **dNBR** raster composite:

```python
# calculate burnt area (pixel size 20m*20m)
reclass = reclassify(dnbr)
k = ['Unburned hectares', 'Low severity hectares', 'Moderate-low severity hectares', 'Moderate-high severity hectares', 'High severity']
for i in range(1,6):
    x = reclass[reclass == i]
    l= x.size*0.04
    print("%s: %.2f" % (k[i-1], l))
```
